```python
# Import packages
import folium
import pandas as pd
from IPython.display import display
```

```python
# Create a DataFrame with fake fire station data

data = pd.DataFrame({
    "Station":["Sation 1", "Sation 2", "Sation 3", "Sation 4", "Sation 5"],
    "Address": ["123 Main St", "456 Park Ave", "789 Elm St", "321 Oak St", " 159 Maple St"],
    "Latitude": [36.1615, 36.1234, 36.1928, 36.1457, 36.1792],
    "Longitude": [-86.7845, -86.7012, -86.7592, -86.7741, -86.7095]
})
```

```python
import folium
import geopy.distance
```

```python
# Create a map with the geospatial data

map = folium.Map(location=[33.1667, -86.7833], zoom_start= 10, tiles = "OpenStreetMap")

# Add the data to the map as markers
for lat, lng, Station in zip (data.Latitude, data.Longitude, data.Station):
    folium.Marker(
        location = [lat, lng],
        icon = folium.Icon(color = "red", icon = "Fire-extinguisher"),
        popup = Station
    ).add_to(map)
# Display the map in the JN
display(map)
```

```python
from itertools import combinations
from geopy.distance import geodesic
```

```python
# Now we are gonna find the distance between these location

for station_1, station_2 in combinations(data.index,2):
    lat1, lng1 = data.loc[station_1, ["Latitude", "Longitude"]]
    lat2, lng2 = data.loc[station_2, ["Latitude", "Longitude"]]
    
    distance = geopy.distance.geodesic((lat1, lng1), (lat2, lng2)).kilometers
# Add a polyline to the map to represent the measured distance
    folium.PolyLine(
        locations = [(lat1, lng1), (lat2, lng2)],
        weight = 2,
        label = "Distance",
        color = "red"
    ). add_to(map)
    
# Display the map in the JN
display(map)
```

```python
# Save the map to an HTML file
map.save("my_map.html")
```
